```python
# Fill missing values in 'PM2.5' with the mean
train['PM2.5'] = train['PM2.5'].fillna(train['PM2.5'].mean())

# Fill missing values in 'SO2' with the mean
train['SO2'] = train['SO2'].fillna(train['SO2'].mean())
```

```python
train['NO2'] = train['NO2'].fillna(train['NO2'].mean())
train['CO'] = train['CO'].fillna(train['CO'].mean())
train['O3'] = train['O3'].fillna(train['O3'].mean())
```

```python
train['TEMP'] =  train['TEMP'].fillna(train['TEMP'].mean())
train['PRES'] =  train['PRES'].fillna(train['PRES'].mean())
train['DEWP'] =  train['DEWP'].fillna(train['DEWP'].mean())
train['RAIN'] =  train['RAIN'].fillna(train['RAIN'].mean())
```

```python
train['wd'].value_counts()
```

```python
mode_wd = train['wd'].mode()[0]

# Fill missing values in 'wd' with the mode
train['wd'] = train['wd'].fillna(mode_wd)
```

```python
train['WSPM'] = train['WSPM'].fillna(train['WSPM'].mean())
```

```python
train.isna().sum()
```

```python
train = pd.get_dummies(train, columns = ['wd'], dtype = 'int')
```

```python
train.head()
```

```python
X = train.drop(['PM10','DateTime','Date','No'], axis=1)
X
```

```python
y = train['PM10']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Multiple linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predicted_x_test = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, predicted_x_test)
mse = mean_squared_error(y_test, predicted_x_test)
rsquare = r2_score(y_test, predicted_x_test)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R Square: {rsquare}")
```

# Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import cross_val_score
```

```python
regressor = DecisionTreeRegressor()
```

```python
cross_val_score(regressor, X_train, y_train, cv=10)
```

```python
regressor.fit(X_train, y_train)
```

```python
ypred=regressor.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, ypred)
mse = mean_squared_error(y_test,ypred )
rsquare = r2_score(y_test, ypred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R Square: {rsquare}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=12)
```

```python
rfr.fit(X_train, y_train)
ypred1=rfr.predict(X_test)

mae = mean_absolute_error(y_test, ypred1)
mse = mean_squared_error(y_test,ypred1 )
rsquare = r2_score(y_test, ypred1)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R Square: {rsquare}")
```
